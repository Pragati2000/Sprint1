# UNDERSTANDING VARIABLE

**1. State/UTs:** The names of Indian states or Union Territories (UTs) where COVID-19 cases were reported. This column allows for geographic analysis of the pandemic across different regions in India.

**2. Total Cases:**: The total number of confirmed COVID-19 cases reported in each state/UT. This figure includes all cases, active or resolved (either recovered or deceased).

**3. Active:** The current number of active COVID-19 cases in each state/UT. These are cases that are ongoing and have not yet resulted in recovery or death.

**4. Discharged:** The total number of patients who have recovered and been discharged from medical care. This represents the number of resolved cases where the patients recovered.

**5.Deaths:** The total number of COVID-19 related deaths in each state/UT. This column tracks fatalities resulting from the virus.

**6.Active Ratio:** This is the ratio of active cases to total cases, expressed as a percentage. It helps to understand the proportion of ongoing infections in comparison to the overall number of cases.

**7.Discharge Ratio:** The percentage of total cases that have resulted in recovery and discharge from medical care.

**8.Death Ratio:** The percentage of total cases that resulted in death. This provides insight into the fatality rate of the virus.

**9. Population:** The total population of each state/UT. This column is crucial for analyzing per capita statistics and understanding the relative impact of the virus in different regions based on their population size.
